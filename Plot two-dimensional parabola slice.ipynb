```python
import tensorflow as tf

devices = tf.config.experimental.get_visible_devices('GPU')
if len(devices) > 0:
    tf.config.experimental.set_memory_growth(devices[0], True)
```

```python
import harmonic_distance as hd
import numpy as np
import matplotlib.pyplot as plt
```

```python
C = 0.1
c = tf.Variable(C, dtype=tf.float64)
```

```python
vs = hd.vectors.VectorSpace(prime_limits=[5, 3, 2], dimensions=2)
```

```python
@tf.function
def loss(x, c=C):
    return hd.optimize.parabolic_loss_function(vs.pds, vs.hds, x, curves=(c))

exp = np.log2([1.0, 6, 2, 3, 4]) * 2
res = loss(np.log2([[1.0, 1.0], [1.5, 1.0], [2.0, 1.0], [3.0, 1.0], [4.0, 1.0]]), c=(0.01, 0.01))
np.testing.assert_equal(exp, res)
res
```

```python
c.assign(0.1)
xs = np.linspace(0.0, 1.0, 2048)
ys = loss(tf.constant(tf.stack([xs, tf.fill(xs.shape, tf.constant(3/2, dtype=tf.float64))], axis=1)), c=c)
plt.plot(xs, ys)
plt.show()
```
